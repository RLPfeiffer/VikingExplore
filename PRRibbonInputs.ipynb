```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import sqlalchemy as sal
from sqlalchemy import create_engine
```

```python
#Connect to Data Structure for sqlalchemy is: "mssql+pyodbc://OpR-Marc-DB2:1433/RPC1?driver=ODBC+Driver+17+for+SQL+Server" --change RPC1 for other volumes

RPC1 = sal.create_engine("mssql+pyodbc://OpR-Marc-DB2:1433/RPC1?driver=ODBC+Driver+17+for+SQL+Server")
connection = RPC1.connect()

RPC2 = sal.create_engine("mssql+pyodbc://OpR-Marc-DB2:1433/RPC2?driver=ODBC+Driver+17+for+SQL+Server")
connection = RPC2.connect()
```

```python
RPC1PRRibbonInputDF = pd.DataFrame()
RPC2PRRibbonInputDF = pd.DataFrame()
```

```python
RPC1RodBCs= ["822", "933", "1001", "1069", "1232", "1243", "1537", "25001"]
RPC2RodBCs = ["2968", "3467", "3578"]
```

```python
#Pull Data from RPC1
for x in RPC1RodBCs: 
    query = "SET NOCOUNT ON; EXEC RLPfeiffer.RibbonInputs @TargetCell= " + "'" + x + "'"
    RI = pd.read_sql(query, con = RPC1)
    RI['Volume']= 'RPC1'
    RPC1PRRibbonInputDF = RPC1PRRibbonInputDF.append(RI)

RPC1PRRibbonInputDF
```

```python
#Pull Data from RPC2
for x in RPC2RodBCs: 
    query = "SET NOCOUNT ON; EXEC RLPfeiffer.RibbonInputs @TargetCell= " + "'" + x + "'"
    RI = pd.read_sql(query, con = RPC2)
    RI['Volume']= 'RPC2'
    RPC2PRRibbonInputDF = RPC2PRRibbonInputDF.append(RI)

RPC2PRRibbonInputDF
```

```python
#concatinate the 2 datasets
frames = [RPC1PRRibbonInputDF, RPC2PRRibbonInputDF]
PRRibbonInputDF = pd.concat(frames).reset_index()
```

```python
PartnersTotal = PRRibbonInputDF.groupby("Volume")['SourceParentLabel'].value_counts()
print(PartnersTotal)
```

```python
PRRibbonInputDF = PRRibbonInputDF.groupby(["TargetParent", 'Volume'])["SourceParentLabel"].value_counts().to_frame(name = 'PRInputCount').reset_index()
PRRibbonInputDF
```

```python
mean_std_df = PRRibbonInputDF.groupby(['Volume', "SourceParentLabel"])['PRInputCount'].agg(['mean', 'std'])
mean_std_df
```
